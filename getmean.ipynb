```python
from keras.datasets import cifar10
(x_train,y_train),(x_test,y_test)=cifar10.load_data()
meanlist = [[0]*3 for i in range(50000)]
```

```python
for j in range(50000):
    r, g, b = cv2.split(x_train[j])
    meanlist[j][0] = int(r.mean())
    meanlist[j][1] = int(g.mean())
    meanlist[j][2] = int(b.mean())
```
